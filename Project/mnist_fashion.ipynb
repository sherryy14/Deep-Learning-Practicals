```
import tensorflow as tf
mnist=tf.keras.datasets.fashion_mnist #fashion_mnist is a dataset
```

```
(training_images, training_labels), (test_images, test_labels)= mnist.load_data()
```

```
training_images=training_images / 255.0
test_images=test_images / 255.0
```

```
#Architecture step

model= tf.keras.models.Sequential([tf.keras.layers.Flatten(),
                                   tf.keras.layers.Dense(512, activation=tf.nn.relu),
                                   tf.keras.layers.Dense(10, activation=tf.nn.softmax)])
#input_shape is not required bcz flatten will give the input
#Flatten is used to convert 2d or 3d array into 1d
#relu is used for activation fn, and if the number=>0 it returns otherwise make the number 0 if it is <0
#Dense means fully connected and layers of all nodes will be connected with previous and next layer's node
#activation fn uses for sending only important data means convert the linearity into non-linearity                                                                            #512 outputs will appear (it's not fixed) we can change and check it                        #softmax is used for multiclassification
#10 outputs... now softmax will appear 10 numbers and that would be the percentage
```

```
#compilation step
model.compile(optimizer=tf.keras.optimizers.Adam(),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```
model.fit(training_images, training_labels, epochs=10, batch_size=128)

#model.fit is used to train data
```

```
model.evaluate(test_images, test_labels)
#epochs means iterations
```
